```python
import numpy as np
import torch

def embed_text(texts):
    # texts: list[str]
    inputs = processor(text=texts, return_tensors="pt", padding=True, truncation=True)
    inputs = {k: v.to(device) for k, v in inputs.items()}

    with torch.no_grad():
        text_features = model.get_text_features(**inputs)

    text_features = text_features / text_features.norm(dim=-1, keepdim=True)
    return text_features.cpu().numpy()
```

```python
label_to_prompt = {
    "SEM": "a scanning electron microscopy (SEM) micrograph",
    "TEM": "a transmission electron microscopy (TEM) micrograph",
    "Optical": "an optical microscopy micrograph",
    "AFM": "an atomic force microscopy (AFM) image",
    "XRD": "an X-ray diffraction (XRD) pattern image",
    "EBSD": "an EBSD map micrograph",
}
```

```python
import re

def normalize_technique(t):
    t = str(t).strip().lower()

    # common matches
    if "sem" in t or "scanning electron" in t:
        return "SEM"
    if "tem" in t or "transmission electron" in t:
        return "TEM"
    if "optical" in t or "light microscopy" in t or "optical microscopy" in t:
        return "Optical"
    if "afm" in t or "atomic force" in t:
        return "AFM"
    if "xrd" in t or "diffraction" in t:
        return "XRD"
    if "ebsd" in t:
        return "EBSD"

    # fallback: keep original compact token
    # (you can also return None to drop unknowns)
    return None
```

```python
df_valid = df_valid.copy()
df_valid["tech_norm"] = df_valid["technique"].apply(normalize_technique)

# keep only techniques we can evaluate
df_eval = df_valid.dropna(subset=["tech_norm"]).reset_index(drop=True)
X_eval = X_img[df_eval.index.values]  # aligned because df_valid order == X_img order
print("Evaluating N =", len(df_eval))
print(df_eval["tech_norm"].value_counts())
```

```python
import torch
from transformers import CLIPProcessor, CLIPModel

device = "cuda" if torch.cuda.is_available() else "cpu"

model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").to(device)
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

model.eval()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Candidate labels present in your eval set
labels = sorted(df_eval["technique"].unique())

# Build prompts in same order
prompts = [label_to_prompt.get(l, f"a {l} microscopy image") for l in labels]

# Text embeddings: shape (num_labels, 512)
T = embed_text(prompts)

# Similarity: (N, num_labels)
S = cosine_similarity(X_eval, T)

pred_idx = S.argmax(axis=1)
pred_labels = [labels[i] for i in pred_idx]

df_eval["pred"] = pred_labels
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Candidate labels present in your eval set
labels = sorted(df_eval["tech_norm"].unique())

# Build prompts in same order
prompts = [label_to_prompt.get(l, f"a {l} microscopy image") for l in labels]

# Text embeddings: shape (num_labels, 512)
T = embed_text(prompts)

# Similarity: (N, num_labels)
S = cosine_similarity(X_eval, T)

pred_idx = S.argmax(axis=1)
pred_labels = [labels[i] for i in pred_idx]

df_eval["pred"] = pred_labels
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

y_true = df_eval["tech_norm"].values
y_pred = df_eval["pred"].values

acc = accuracy_score(y_true, y_pred)
print("Zero-shot technique accuracy:", round(acc, 4))
print()

print(classification_report(y_true, y_pred, labels=labels))
```

```python
import matplotlib.pyplot as plt
import numpy as np

cm = confusion_matrix(y_true, y_pred, labels=labels)

plt.figure(figsize=(6.5, 5.5))
plt.imshow(cm, interpolation="nearest")
plt.title("Zero-shot Technique Confusion Matrix (CLIP)")
plt.xticks(range(len(labels)), labels, rotation=45, ha="right")
plt.yticks(range(len(labels)), labels)
plt.xlabel("Predicted")
plt.ylabel("True")

# annotate counts
for i in range(len(labels)):
    for j in range(len(labels)):
        plt.text(j, i, cm[i, j], ha="center", va="center", fontsize=9)

plt.tight_layout()
plt.show()
```

```python
top2 = np.argsort(S, axis=1)[:, -2:]
top2_labels = [[labels[i] for i in row] for row in top2]

top2_correct = np.mean([yt in preds for yt, preds in zip(y_true, top2_labels)])
print("Top-2 accuracy:", round(top2_correct, 4))
```

```python
# ------------------------------------
# Prepare multi-label ground truth
# ------------------------------------
df_eval["cat_list"] = (
    df_eval["categories"]
    .fillna("")
    .str.split(",")
    .apply(lambda xs: [x.strip() for x in xs if x.strip() != ""])
)
```

```python
from collections import Counter

all_cats = [c for cats in df_eval["cat_list"] for c in cats]
cat_counts = Counter(all_cats)

TOP_K = 10
cat_labels = [c for c, _ in cat_counts.most_common(TOP_K)]

print(cat_labels)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer(classes=cat_labels)
Y_true_cat = mlb.fit_transform(df_eval["cat_list"])

print(Y_true_cat.shape)   # (N_images, TOP_K)
```

```python
cat_prompts = [
    f"a materials science micrograph of {c}, showing its microstructure"
    for c in cat_labels
]

```

```python
with torch.no_grad():
    text_inputs = processor(
        text=cat_prompts,
        return_tensors="pt",
        padding=True
    ).to(device)

    text_features = model.get_text_features(**text_inputs)
    text_features = text_features / text_features.norm(dim=1, keepdim=True)

X_img = np.load("clip_image_embeddings.npy")   # (867, 512)

image_features = torch.tensor(X_img).to(device)
image_features = image_features / image_features.norm(dim=1, keepdim=True)

with torch.no_grad():
    S_cat = (image_features @ text_features.T).cpu().numpy()
```

```python
Y_pred_cat = np.zeros_like(Y_true_cat)

top2 = np.argsort(S_cat, axis=1)[:, -2:]
for i in range(Y_pred_cat.shape[0]):
    Y_pred_cat[i, top2[i]] = 1
```

```python
top2_cat_correct = np.mean([
    np.any(Y_true_cat[i, top2[i]] == 1)
    for i in range(Y_true_cat.shape[0])
])

print("Top-2 category accuracy:", round(top2_cat_correct, 4))
```

```python
top1 = np.argmax(S_cat, axis=1)

top1_cat_acc = np.mean([
    Y_true_cat[i, top1[i]] == 1
    for i in range(Y_true_cat.shape[0])
])

print("Top-1 category accuracy (Hit@1):", round(top1_cat_acc, 4))
```

```python
query = "a scanning electron microscopy image of a ceramic"

text_inputs = processor(
    text=[query],
    return_tensors="pt",
    padding=True
).to(device)

with torch.no_grad():
    query_feat = model.get_text_features(**text_inputs)
    query_feat = query_feat / query_feat.norm(dim=1, keepdim=True)
```

```python
scores = (image_features @ query_feat.T).squeeze().cpu().numpy()

top = np.argsort(scores)[-5:][::-1]   # top-5 most similar
```
